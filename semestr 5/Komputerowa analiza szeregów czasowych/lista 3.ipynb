```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats
```

### Zadanie 1

```python
beta0 = 2
beta1 = 5
ns = np.arange(5, 101, 5)
sigmas = [0.01, 0.5, 1]
alfa = 0.05
for sigma in sigmas:
    beta0Per = []
    beta1Per = []
    for n in ns:
        beta0In = 0
        beta1In = 0
        for m in range(500):
            x = np.linspace(1,n,n)
            x_mean = np.mean(x)
            ei = np.random.normal(0,sigma,size = n)
            y = beta0+beta1*x+ei
            y_mean = np.mean(y)
            beta1_hat = np.sum((x-x_mean)*(y-y_mean))/np.sum((x-x_mean)**2)
            beta0_hat = y_mean-beta1_hat*x_mean
            leftBeta0 = beta0_hat-scipy.stats.norm.ppf(1-alfa/2)*sigma*np.sqrt(1/n + x_mean**2/np.sum((x-x_mean)**2))
            rightBeta0 = beta0_hat+scipy.stats.norm.ppf(1-alfa/2)*sigma*np.sqrt(1/n + x_mean**2/np.sum((x-x_mean)**2))
            leftBeta1 = beta1_hat - scipy.stats.norm.ppf(1-alfa/2)*sigma/np.sqrt(np.sum((x-x_mean)**2))
            rightBeta1 = beta1_hat + scipy.stats.norm.ppf(1-alfa/2)*sigma/np.sqrt(np.sum((x-x_mean)**2))
            if beta0 >= leftBeta0 and beta0 <= rightBeta0:
                beta0In += 1
            if beta1 >= leftBeta1 and beta1 <= rightBeta1:
                beta1In += 1
        beta0Per.append(beta0In/500)
        beta1Per.append(beta1In/500)
    plt.plot(ns, beta0Per)
    plt.plot(ns, beta1Per)
    plt.show()
    plt.clf()

```

```python
beta0 = 2
beta1 = 5
ns = np.arange(5, 101, 5)
sigmas = [0.01, 0.5, 1]
alfa = 0.05
for sigma in sigmas:
    beta0Per = []
    beta1Per = []
    for n in ns:
        beta0In = 0
        beta1In = 0
        for m in range(500):
            x = np.linspace(1,n,n)
            x_mean = np.mean(x)
            ei = np.random.normal(0,sigma,size = n)
            y = beta0+beta1*x+ei
            y_mean = np.mean(y)
            beta1_hat = np.sum((x-x_mean)*(y-y_mean))/np.sum((x-x_mean)**2)
            beta0_hat = y_mean-beta1_hat*x_mean
            yHat = beta0_hat+beta1_hat*x
            s = np.sqrt(np.sum((yHat-y)**2)/(n-2))
            leftBeta0 = beta0_hat-scipy.stats.t.ppf(1-alfa/2, df=n-2)*s*np.sqrt(1/n + x_mean**2/np.sum((x-x_mean)**2))
            rightBeta0 = beta0_hat+scipy.stats.t.ppf(1-alfa/2, df=n-2)*s*np.sqrt(1/n + x_mean**2/np.sum((x-x_mean)**2))
            leftBeta1 = beta1_hat - scipy.stats.t.ppf(1-alfa/2, df=n-2)*s/np.sqrt(np.sum((x-x_mean)**2))
            rightBeta1 = beta1_hat + scipy.stats.t.ppf(1-alfa/2, df=n-2)*s/np.sqrt(np.sum((x-x_mean)**2))
            if beta0 >= leftBeta0 and beta0 <= rightBeta0:
                beta0In += 1
            if beta1 >= leftBeta1 and beta1 <= rightBeta1:
                beta1In += 1
        beta0Per.append(beta0In/500)
        beta1Per.append(beta1In/500)
    plt.plot(ns, beta0Per)
    plt.plot(ns, beta1Per)
    plt.show()
    plt.clf()
```

### Zadanie 3

```python
n = 100
alfa = 0.05
beta0 = 2
beta1 = 5
xs = np.linspace(1,n,n)
xMean = np.mean(xs)
gammas = np.arange(-1,1.1,0.1)
sigma = 1
gamma_left1 = []
gamma_right1 = []
x0s = []
for gamma in gammas:
    x0 = xMean+gamma
    ei = np.random.normal(0,sigma,size = n)
    y = beta0+beta1*xs+ei
    yMean = np.mean(y)
    beta1Hat = np.sum((xs-xMean)*(y-yMean))/np.sum((xs-xMean)**2)
    beta0Hat = yMean-beta1Hat*xMean
    left = beta0Hat + beta1Hat*x0 - scipy.stats.norm.ppf(1-alfa/2)*np.sqrt(sigma**2 * (1/n + ((x0-xMean)**2)/np.sum((xs-xMean)**2)))
    right = beta0Hat + beta1Hat*x0 + scipy.stats.norm.ppf(1-alfa/2)*np.sqrt(sigma**2 * (1/n + ((x0-xMean)**2)/np.sum((xs-xMean)**2)))
    gamma_left1.append(left)
    gamma_right1.append(right)
    x0s.append(x0)
plt.plot(gammas, gamma_left1)
plt.plot(gammas, gamma_right1)
```
