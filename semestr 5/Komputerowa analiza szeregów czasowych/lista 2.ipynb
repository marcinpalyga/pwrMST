```python
n = 1000
beta0 = 2
beta1 = 5
sigma = 2
x = np.linspace(0,1,n)
x_mean = np.mean(x)
betas0 = []
betas1 = []
ws0 = []
ws1 = []
for i in range(1000):
    e_i = np.random.normal(0, sigma, size=n)
    y = beta0 + beta1*x+e_i
    y_mean = np.mean(y)
    beta1_hat = np.sum((x-x_mean)*(y-y_mean))/np.sum((x-x_mean)**2)
    beta0_hat = y_mean-beta1_hat*x_mean
    yHat = beta0_hat+beta1_hat*x
    s = np.sqrt(np.sum((yHat-y)**2)/(n-2))
    seBeta0 =  s*np.sqrt((x_mean)**2/(np.sum((x-x_mean)**2))+1/n)
    seBeta1 = s/np.sqrt(np.sum((x-x_mean)**2))
    ws0.append((beta0_hat-beta0)/seBeta0)
    ws1.append((beta1_hat-beta1)/seBeta1)
    
```

```python
count, bins, ignored = plt.hist(ws0, bins='sqrt', density=True)
x = np.linspace(scipy.stats.t.ppf(0.001, n-2),
                scipy.stats.t.ppf(0.999, n-2), 100)
plt.plot(x, scipy.stats.t.pdf(x, n-2),
       'r-')
plt.show()
```
